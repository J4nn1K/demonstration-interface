```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib widget

import tikzplotlib


cphs = [16,13,11]
cph = np.round(np.mean(cphs)*20)


# Data for the first y-axis
categories = ['Human Hand', 'Handheld Gripper', 'Teleoperation']
values1 = [34*20, cph, 0]

# Data for the second y-axis
values2 = [231, 111, 35]

fig, ax1 = plt.subplots()

# Plotting the first y-axis
bar_width = 0.4
x = np.arange(len(categories))
bars1 = ax1.bar(x - bar_width/2, values1, bar_width, label='Ours', color='tab:blue')

# Setting labels and title for the first y-axis
# ax1.set_xlabel('Categories')
ax1.set_ylabel('Collections per Hour (Ours)')
# ax1.set_title('Bar Chart with Two Different Y-Axes')
ax1.tick_params(axis='y')
ax1.set_xticks(x)
ax1.set_xticklabels(categories)

# Adding values on top of the first set of bars
for bar in bars1:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), ha='center', va='bottom')  # va: vertical alignment

# Creating a second y-axis
ax2 = ax1.twinx()
bars2 = ax2.bar(x + bar_width/2, values2, bar_width, label='UMI', color='tab:orange')

# Setting labels for the second y-axis
ax2.set_ylabel('Collections per Hour (UMI)')
ax2.tick_params(axis='y')

for bar in bars2:
    yval = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), ha='center', va='bottom')  # va: vertical alignment


# Adding legend
fig.legend(loc='upper right', bbox_to_anchor=(0.85, 0.85))

# Displaying the plot
#plt.show()
tikzplotlib.clean_figure()
tikzplotlib.save("throughput.tex")
```
